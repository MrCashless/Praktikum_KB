### Import Library yang akan digunakan

```python
import pandas as pd
import numpy as np
```

### Membaca data file CSV

```python
df = pd.read_csv("FIFA_data.csv")
```

### Mengetahui informasi dari isi data file CSV

```python
df.info()
```

### Mengimpor data file CSV ke dalam Dictionary

```python
df_show = {
'atribut': [],
'atribut_objek': 0,
'atribut_nomor': 0,
'atribut_null': 0,
'jml_record': 0,
}
```

```python
df_show['atribut'] = list(df.columns.values)
df_show['atribut_objek'] = list(df.select_dtypes(['object']).columns)
df_show['atribut_nomor'] = list(df.select_dtypes(['int64','float64']).columns)
df_show['atribut_null'] = list(df.loc[:, df.isnull().any()].columns)
df_show['jml_record'] = df.shape[0]
```

### Menampilkan Hasil dict df_show:

### A) Jumlah Record Data

```python
print("Jumlah Record Pada Data :", df_show['jml_record'])
```

### B) Nama Attributes Data

```python
print("Nama Seluruh Atribut Data :", df_show['atribut'])
```

### C) Jumlah Attributes Data

```python
print("Total Atribut Data :", len(df_show['atribut']))
print("Total Atribut Dengan Tipe Data Object :", len(df_show['atribut_objek']))
print("Total Atribut Dengan Tipe Data Angka :", len(df_show['atribut_nomor']))
print("Total Atribut Null :", len(df_show['atribut_null']))
```
